# Highly divisible triangular number

The sequence of triangle numbers is generated by adding the natural numbers. 
So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. 

The first ten terms would be:
```
1, 3, 6, 10, 15, 21, 28, 36, 45, 55, ...
```

Let us list the factors of the first seven triangle numbers:

```
 1: 1
 3: 1,3      
 6: 1,2,3,6
10: 1,2,5,10
15: 1,3,5,15
21: 1,3,7,21
28: 1,2,4,7,14,28
```

We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?

````python
def divisors(n):
    """ Return a list of divisors of positive int n

    Parameters
    ----------
    n (int): A number to be divided

    Return
    ------
    divisors (list of int): the divisors """

    
    if not isinstance(n, int) or n < 1:
        return []

    divisors = set([1, n])

    for i in range(2, int(n / 2)):
        if n % i == 0:
            divisors.add(i)
            divisors.add(int(n / i))

    return list(divisors)

divisors(6)
````

````python
n_tri = 0
n_div_max = 0
n_div_limit = 500

for i in range(1, int(1e+20)):
    n_tri += i
    n_div = len(divisors(n_tri))
    if n_div > n_div_max:
        n_div_max = n_div
        print("i = {}, n_tri = {}, n_div = {}".format(i, n_tri, n_div))
        # stop when n_div is over n_div_max
        if n_div_max > n_div_limit:
            print("The first triangle number to have over {} divisors = {} (n_div = {})".format(n_div_limit, n_tri, n_div_max))
            break

    
````

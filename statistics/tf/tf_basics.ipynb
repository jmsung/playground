# auto diff

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    y = x**2 + x + 1
    return y

x = tf.Variable(np.linspace(-10, 10, 100))

with tf.GradientTape() as tape:
  y = f(x)

g_x = tape.gradient(y, x)  # g(x) = dy/dx

plt.plot(x, y, 'k')
plt.plot(x, g_x, 'r')
plt.show()
```
